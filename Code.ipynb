### IGSO Algorithm

```python
#updates step size according to change in f
def updatestep(change, step, threshold):
    if(change > threshold): step *= 1.2
    else: step *= 0.8
    return step
```

```python
#calculates value of f to be optimized
def function(x, y):
    func = (x-2)**2 + (y-3)**2  #this is the established function (to find minimum of)
    return func
```

```python
#initiates neighbors for every new chosen node / point
def gridSearch(x, y, step, threshold):
    currentF = function(x, y)
    print("currect F value: ", currentF, "\n")

    newF = currentF

    xi = [x - step, x, x + step]
    yi = [y - step, y, y + step]

    #finds the function with lowest f value among all neighbors
    for i in xi:
        for j in yi:
            checkF = function(i, j)
            if(checkF < currentF):
                newF = checkF
                x = i
                y = j

            print(i, j, checkF)

    #updates step according to delta f between new node and current node
    print("new F value: ", newF)
    change = abs(currentF - newF)
    step = updatestep(change, step, threshold)

    return x, y, step
```

###Parameter Initialization

```python
threshold = 0.1
step = 0.5
x = 5
y = 5
```

###Running Example

```python
iteration = 1
F = function(x, y)

while F > 0.01 and step > 0.01:
    print("\niteration", iteration)
    print("X:", x, " Y:", y, " step:", step, " Function:", F)

    x, y, step = gridSearch(x, y, step, threshold)

    F = function(x, y)

    iteration += 1

print("\nX:", x, " Y:", y, " step:", step, " Function:", F)
```
